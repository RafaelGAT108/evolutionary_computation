# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Execute Functions

```python
def yperceptron(W: np.ndarray, b: np.ndarray, X: np.ndarray):
    '''
        W (np.ndarray): represent the weights of the neuron
        b (np.ndarray): represent the bias of the neuron
        X (np.ndarray): represent the inputs of the neuron
    '''
    result = np.dot(W, X) + b
    return [1 if v >= 0 else 0 for v in result[0]]
```

# AND Execution

```python
W = np.array([[0.4, 0.4]])
b = -.6
X = np.array([[0, 1, 0, 1], [0, 0, 1, 1] ])

print(f'W shape: {W.shape}, X shape: {X.shape}')

x1 = X[0] 
x2 = X[1] 

y = yperceptron(W, b, X)
print(f'Y = {y}')

for i in range(len(y)):
    if x1[i] == 1 and x2[i] == 1:
        plt.plot(x1[i], x2[i], 'r+', label='+: y = 1' if y[i] == 1 else None)
    else:
        plt.plot(x1[i], x2[i], 'bo', label='o: y = 0' if y[i] == 0 else None)

handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())
plt.title("y results for AND logic")
plt.xlabel('X1')
plt.ylabel('X2')

plt.grid(True)
plt.show()
```

# OR Execution

```python
W = np.array([[0.4, 0.4]])*(1/.1)
b = -.6
X = np.array([[0, 1, 0, 1], [0, 0, 1, 1] ])
x1 = X[0] 
x2 = X[1] 
print(f'W shape: {W.shape}, X shape: {X.shape}')

y = yperceptron(W, b, X)
print(f'Y = {y}')


for i in range(len(y)):
    if x1[i] == 0 and x2[i] == 0:
        plt.plot(x1[i], x2[i], 'bo', label='+: y = 0' if y[i] == 0 else None)
    else:
        plt.plot(x1[i], x2[i], 'r+', label='o: y = 1' if y[i] == 1 else None)

handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())
plt.title("y results for OR logic")
plt.xlabel('X1')
plt.ylabel('X2')

plt.grid(True)
plt.show()
```
